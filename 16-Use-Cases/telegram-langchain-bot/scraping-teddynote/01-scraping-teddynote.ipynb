# <랭체인LangChain 노트> 스크래핑 하기

````python
# 필요한 라이브러리 임포트
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time

# Chrome 브라우저 옵션 설정 (헤드리스 모드로 실행)
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # 브라우저 창을 띄우지 않고 실행

# Chrome 드라이버 서비스 생성
service = Service(ChromeDriverManager().install())

# 웹 드라이버 초기화
driver = webdriver.Chrome(service=service, options=options)

# 테디노트 랭체인(Langchain-KR) 사이트 URL 설정
ebook ="https://wikidocs.net/book/14314"
base_url = "https://wikidocs.net"

# 설정된 URL로 접속
driver.get(ebook)

# 페이지 소스 가져오기 위해 BeautifulSoup 사용
from bs4 import BeautifulSoup
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'html.parser')
````

````python
# 목차 카테고리: 'a' 태그 선택
toc = soup.select('body > div > div.content.row > div.col-sm-3 > div > div > div.toc.toc-checker > div.list-group.list-group-toc > a')
````

````python
# 추출한 데이터 저장할 리스트
data_list = []

# 'a' 태그를 순회하며 title과 sub_url 추출
for item in toc:
    span_tag = item.find('span')
    title = span_tag.get('title') if span_tag else item.get('title')
    sub_url = item.get('href').split('page(')[-1].rstrip(')')
    if not sub_url:
        sub_url = item.get('href').split('/')[-1]
    full_url = f"{base_url}/{sub_url}"
    data_list.append({"title": title, "source": full_url})

# 출력
for data in data_list:
    print(f"title: {data['title']}, source: {data['source']}")
````

# Source link를 순회하며 Contents 스크래핑

````python
# 주어진 URL에서 콘텐츠를 추출하고 출력하는 함수를 정의
def extract_content(link):
    # 지정된 링크로 접속
    driver.get(link)
    time.sleep(3)  # 페이지 로딩 시간을 위해 잠시 대기

    # 페이지 소스를 가져와서 BeautifulSoup으로 파싱
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')

    # 제목 추출
    title_tag = soup.find('h1', class_='page-subject')
    title = title_tag.get_text(strip=True) if title_tag else "No title found"

    # 주요 콘텐츠 추출
    main_content = soup.find('div', class_='page-content tex2jax_process')
    if not main_content:
        print("No main content found for", link)
        return

    # 관련 텍스트 모두 추출
    extracted_text = [f"# {title}\n"]

    for tag in main_content.find_all(['h2', 'p', 'ul', 'li', 'pre']):
        if tag.name == 'pre':
            extracted_text.append(f"```\n{tag.get_text(strip=True)}\n```")
        elif tag.name == 'h2':
            extracted_text.append(f"## {tag.get_text(strip=True)}")
        elif tag.name == 'ul':
            for li in tag.find_all('li'):
                extracted_text.append(f"- {li.get_text(strip=True)}")
        else:
            extracted_text.append(tag.get_text(strip=True))

    # 추출한 텍스트를 하나의 문자열로 결합
    result = '\n'.join(extracted_text)
    return result

# 데이터 리스트의 두 번째 항목부터 순회하면서 콘텐츠 추출
for item in data_list[1:]:
    item['content'] = extract_content(item['source'])
````

````python
import pandas as pd
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain.vectorstores.faiss import FAISS
from langchain.schema import Document  # 이 부분을 추가합니다.

# CSV 파일을 pandas DataFrame으로 로드합니다.
df = pd.read_csv('data_list_with_content.csv')

# 텍스트 분할기 설정
text_splitter = RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=200)

# 문서 리스트를 생성합니다.
documents = []
for index, row in df.iterrows():
    if pd.isna(row['content']):
        continue
    chunks = text_splitter.split_text(row['content'])
    for chunk in chunks:
        documents.append(Document(page_content=chunk, metadata={"title": row['title'], "source": row['source']}))

print(f"분할된 테디노트 파일의 개수: {len(documents)}")

# OpenAI 임베딩 로드
embeddings = OpenAIEmbeddings()

# FAISS 인덱스를 생성합니다.
db = FAISS.from_documents(documents, embeddings)

# FAISS 인덱스를 로컬에 저장합니다.
db.save_local(folder_path="teddy-notes_faiss")
print("FAISS 인덱스가 'teddy-notes_faiss'에 생성 및 저장되었습니다.")

# 검색을 위해 FAISS 인덱스를 로드합니다.
faiss_index = FAISS.load_local("teddy-notes_faiss", embeddings, allow_dangerous_deserialization=True)

def search(query):
    # FAISS 인덱스를 검색합니다.
    results = faiss_index.similarity_search(query, k=5)  # 필요에 따라 k 조정

    return results

print("\n\n")

# 예제 검색
query = "LangChain 사용 방법"
print("[검색질의] :" + query)
print("\n[검색결과] :\n")
results = search(query)
for result in results:
    print(result)
    print("\n\n")
````
